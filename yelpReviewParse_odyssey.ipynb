```python
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
reviews = []
with open('yelp_academic_dataset_review.json', encoding='utf8') as f:
    for line in f:
        reviews.append(json.loads(line))
```

```python
df = json_normalize(reviews)
df.columns = df.columns.map(lambda x: x.split(".")[-1])
df
```

```python
list(df)
```

```python
odyssey_df = pd.read_csv("odyssey_restaurants.csv")
del(odyssey_df['Unnamed: 0'])
```

```python
odyssey_reviews = df.loc[df['business_id'].isin(odyssey_df.business_id)]
len(odyssey_df.business_id.unique())

len(odyssey_reviews)
```

```python
odyssey_reviews.to_csv("odyssey_reviews.csv")
```
